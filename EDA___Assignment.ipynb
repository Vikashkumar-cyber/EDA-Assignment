Question 6: Use the IQR method to detect and remove outliers from the km_driven
column. Show before-and-after summary statistics.

```python
# Before outlier removal
print("Summary statistics before outlier removal:")
print(df['km_driven'].describe())

Q1 = df['km_driven'].quantile(0.25)
Q3 = df['km_driven'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_no_outliers = df[(df['km_driven'] >= lower_bound) & (df['km_driven'] <= upper_bound)]

# After outlier removal
print("\nSummary statistics after outlier removal:")
print(df_no_outliers['km_driven'].describe())
```

Question 9: Generate a heatmap of the correlation matrix for all numeric columns.
What correlations stand out the most?


```python
# Select only numeric columns for correlation matrix
numeric_df = df.select_dtypes(include=np.number)

plt.figure(figsize=(10, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Numeric Columns')
plt.show()

# Identify and describe the most notable correlations
correlation_matrix = numeric_df.corr()
# We are interested in strong correlations, either positive or negative, excluding the correlation of a variable with itself
# Get the absolute values of the correlations
abs_corr_matrix = correlation_matrix.abs()
# Fill the diagonal with 0s so we don't pick the correlation of a variable with itself
np.fill_diagonal(abs_corr_matrix.values, 0)

# Find the variables with the highest absolute correlation
max_corr = abs_corr_matrix.unstack().sort_values(ascending=False)
# Get the top 2 pairs (since correlation is symmetric, we will get each pair twice)
top_corr_pairs = max_corr[max_corr > 0].head(2)

print("\nMost notable correlations:")
for index, value in top_corr_pairs.items():
    var1, var2 = index
    print(f"- {var1} and {var2}: {correlation_matrix.loc[var1, var2]:.2f}")
```
